**2.One Sample T-Student Test**

```python
from scipy.stats import ttest_1samp
```

```python
T_Statistic, p_value1 = ttest_1samp(df['A'], 39)

print('T-Statistic: %.3f' % T_Statistic)
print('p-value: %.3f' % p_value1)
if p_value1 < 0.05:
    print ("Reject Null Hypothesis")
else:
    print ("Failed to reject Null Hypothesis")
```

```python
T_Statistic, p_value1 = ttest_1samp(df['A'], 100)

print('T-Statistic: %.3f' % T_Statistic)
print('p-value: %.3f' % p_value1)
if p_value1 < 0.05:
    print ("Reject Null Hypothesis")
else:
    print ("Failed to reject Null Hypothesis")
```

**3.Paired T-Student Test**

```python
from scipy.stats import ttest_rel
```

, , 

```python
df1 = pd.DataFrame({"Before":[88, 82, 84, 93, 75, 78, 84, 87, 95, 91, 83, 89, 77, 68, 91],
                    "After":[91, 84, 88, 90, 79, 80, 88, 90, 90, 96, 88, 89, 81, 74, 92]})

df1.head()
```

```python
df1.mean()
```

```python
T_Statistic2, p_value2 = ttest_rel(df1['After'], df1['Before'])

print('T-Statistic: %.3f' % T_Statistic2)
print('p-value: %.3f' % p_value2)
if p_value2 < 0.05:
    print ("Reject Null Hypothesis")
else:
    print ("Failed to reject Null Hypothesis")
```

**Analysis of Variance Test (ANOVA)**

**One-Way ANOVA**

```python
from scipy.stats import f_oneway
```

```python
df2 = pd.DataFrame({"N1":[89, 89, 88, 78, 79], "N2":[93, 92, 94, 89, 88],
                    "N3":[89, 88, 89, 93, 90], "N4":[81, 78, 81, 92, 82]})
df2
```

```python
df2.mean()
```

```python
df2.var()
```

```python
T_Statistic3, p_value3 = f_oneway(df2["N1"], df2["N2"], df2["N3"], df2["N4"])
print('T-Statistic: %.3f' % T_Statistic3)
print('p-value: %.3f' % p_value3)
if p_value3 < 0.05:
    print ("Reject Null Hypothesis")
else:
    print ("Failed to reject Null Hypothesis")
```
